```python
def Scrap_statistic_purpalcap(year):
    driver = webdriver.Chrome()
    url = f"https://www.iplt20.com/stats/{year}"
    driver.get(url)
    driver.maximize_window()
    time.sleep(1)
    driver.find_element(By.XPATH,'/html/body/div[3]/div/button').click()
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, 200);")
    time.sleep(1)
    driver.find_element(By.XPATH,'/html/body/div[2]/div[4]/div/section/div/div[3]/div[2]/div[1]/div/div/div[2]/div/div[1]').click()
    time.sleep(1)
    driver.find_element(By.XPATH,'/html/body/div[2]/div[4]/div/section/div/div[3]/div[2]/div[1]/div/div/div[2]/div/div[2]/div[1]/span[2]').click()
    time.sleep(1)
    driver.find_element(By.XPATH,'/html/body/div[2]/div[4]/div/section/div/div[3]/div[2]/div[1]/div/div/div[2]/div/div[2]/div[15]').click()
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, 2000);")
    time.sleep(5)
    driver.find_element(By.XPATH,'/html/body/div[2]/div[4]/div/section/div/div[3]/div[2]/div[2]/div/div[3]/div/div[2]/div/a').click()
    time.sleep(1)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    time.sleep(1)
    table_class_names = ['st-table statsTable ng-scope archiveseason', 'st-table statsTable ng-scope']
    table = None
    for class_name in table_class_names:
        table = soup.find('table', {'class': class_name})
        if table:
            break
    player_pos=[]
    player_name=[]
    player_mat=[]
    player_inns =[]
    player_ov=[]
    player_runs=[]
    player_wkts=[]
    player_bbi=[]
    player_avg=[]
    player_econ=[]
    player_sr=[]
    player_4w=[]
    player_5w=[]
   
    
    for row in table.find_all('tr'):
        
        cells = row.find_all('td')
        if len(cells) > 0:
            # Example: Extract player name and runs scored
            player_pos.append(cells[0].text.strip())
            player_name.append(cells[1].text.strip())
            player_mat.append(cells[2].text.strip())
            player_inns.append(cells[3].text.strip())
            player_ov.append(cells[4].text.strip())
            player_runs.append(cells[5].text.strip())
            player_wkts.append(cells[6].text.strip())
            player_bbi.append(cells[7].text.strip())
            player_avg.append(cells[8].text.strip())
            player_econ.append(cells[9].text.strip())
            player_sr.append(cells[10].text.strip())
            player_4w.append(cells[11].text.strip())
            player_5w.append(cells[12].text.strip())
    dictionary ={
        'POS':player_pos,
        'Player':player_name,
        'MAT': player_mat,
        'INNS':player_inns,
        'OV':player_ov,
        'Runs taken':player_runs,
        'WKTS':player_wkts,
        'BBI':player_bbi,
        'AVG WKT':player_avg,
        'ECON':player_econ,
        'SR':player_sr,
        '4W':player_4s,
        '5W':player_5w
    }
    df=pd.DataFrame(dictionary)
    return df
        
    
    
```
