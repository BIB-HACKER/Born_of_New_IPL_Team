```python
# Connect to IPL database
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="Mydatabase@1234",
    database="IPL"
)
cursor = conn.cursor()
```

```python
# Create player table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Players (
    PlayerID INT AUTO_INCREMENT PRIMARY KEY,
    PlayerName VARCHAR(255),
    TeamName VARCHAR(255)
)
""")
print("Player table created successfully")
```

```python
players = pd.read_csv('players.csv')
players
```

```python
# Insert unique player data into the Players table
for index, row in players.iterrows():
    cursor.execute("INSERT INTO players (PlayerName, TeamName) VALUES (%s, %s)", (row['Player'], row['Team']))
print("Data inserted successfully")
```

```python
#Commit changes and close connection
conn.commit()
conn.close()
```
