```python
from sqlalchemy import create_engine

# connection string
connection_string = 'mysql+pymysql://root:Mydatabase1234@127.0.0.1:3306/cricket'

```

```python
# Create SQLAlchemy engine
engine = create_engine(connection_string)
```

```python
playerInfo_df = pd.read_excel('playerinfo.xlsx')
```

```python
playerInfo_df = playerInfo_df.rename(columns={'Player':'Name', 'playerId': 'PlayerID'})
```

```python
playerInfo_df=playerInfo_df[['PlayerID', 'Name', 'Type', 'Nationality', 'Team']]
```

```python
playerInfo_df.head()
```

```python
# Create player table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Players (
    PlayerID INT,
    Name VARCHAR(255),
    Type VARCHAR(100),
    Nationality VARCHAR(200),
    Team VARCHAR(100)
    
)
""")
print("Players table created successfully")
```

```python

playerInfo_df.to_sql('Players', con=engine, if_exists='append', index=False)

# Close the database connection
#engine.dispose()
```
