```python
def Scrap_playes_statistic(year):
    driver = webdriver.Chrome()
    url = f"https://www.iplt20.com/stats/{year}"
    driver.get(url)
    driver.maximize_window()
    driver.find_element(By.XPATH,'/html/body/div[3]/div/button').click()
    driver.execute_script("window.scrollTo(0, 2400);")
    time.sleep(2)
    driver.find_element(By.XPATH,'//*[@id="battingTAB"]/div/a').click()
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup.find('table',{'class':'st-table statsTable ng-scope archiveseason'})
    player_pos=[]
    player_name=[]
    player_mat=[]
    player_inns=[]
    player_no=[]
    player_runs=[]
    player_hs=[]
    player_avg=[]
    player_bf=[]
    player_sr=[]
    player_100=[]
    player_50=[]
    player_4s=[]
    player_6s=[]
    
    for row in table.find_all('tr'):
        
        cells = row.find_all('td')
        if len(cells) > 0:
            # Example: Extract player name and runs scored
            player_pos.append(cells[0].text.strip())
            player_name.append(cells[1].text.strip())
            player_mat.append(cells[2].text.strip())
            player_inns.append(cells[3].text.strip())
            player_no.append(cells[4].text.strip())
            player_runs.append(cells[5].text.strip())
            player_hs.append(cells[6].text.strip())
            player_avg.append(cells[7].text.strip())
            player_bf.append(cells[8].text.strip())
            player_sr.append(cells[9].text.strip())
            player_100.append(cells[10].text.strip())
            player_50.append(cells[11].text.strip())
            player_4s.append(cells[12].text.strip())
            player_6s.append(cells[13].text.strip())
    dictionary ={
        'POS':player_pos,
        'Player':player_name,
        'MAT': player_mat,
        'INNS':player_inns,
        'NO':player_no,
        'RUNS':player_runs,
        'HS':player_hs,
        'AVG':player_avg,
        'BF':player_bf,
        'SR':player_sr,
        '_100s':player_100,
        '_50s':player_50,
        '_4S':player_4s,
        '_6S':player_6s
    }
    df=pd.DataFrame(dictionary)
    return df
        
    
    
```
